```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
StudyHrs = np.array([1,2,3,4,5,6]) # independent Variable
Marks = np.array([85,92,88,95,90,93]) # dependent Variable
```

```python
df = pd.DataFrame({
    'study Hours': StudyHrs,
    'Marks': Marks
})
df
```

```python
# For Better Visualation Let's Plot
plt.plot(df['study Hours'],df['Marks'])
plt.title('Study Hours vs Marks')
plt.xlabel('Study Hours')
plt.ylabel('Marks')
plt.show()

# Now, let's Create a Linear Regression model
```

```python
# let create Ml Model Which Create Best Fit Line so That, We Can Predit Somthiong
```

```python
model = LinearRegression() # Assign The ML Model(Algo/Math) in "Model" Variable
```

```python
# Let Train The Model
model.fit(df[['study Hours']],df['Marks']) # This will Learn The Cofficients(M/Slope)
```

```python
# Now The Model Is Trained On Previous Data
# It Can Predict Something New 
```

```python
model.predict([[3.5]])
```
